# Análisis de las Variables (Graficos)

```python


import matplotlib.pyplot as plt
```

```python
f,ax=plt.subplots(figsize = (18,18))
sns.heatmap(df_cancer.corr(),annot= True,linewidths=0.5,fmt = ".1f",ax=ax)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.title('Mapa de correlación de toda la base')
plt.show()
```

### Interpretacion
El grafico nos muestra la correlacion que hay entre las variables tanto para cancer maligno y benigno. Nos muestra relevancia entre: el radio promedio y area,perimetro,radio del cancer; textura media y textura del cancer; perimetro medio y area, perimetro, radio del cancer; area media y area, perimetro, radio del cancer; Además de una correlacion entre: el radio promedio y area, perimetro medio; perimetro medio y area media.
